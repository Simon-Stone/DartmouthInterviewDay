```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import librosa
import librosa.display
```

```python
# Load wave file
y, sr = librosa.load("wav/vowels.wav", sr=None)
```

```python
# Plot and save /i/ waveform
plt.figure(figsize=(6, 4), dpi=300)
plt.plot(y[int(0.25*sr):int(0.35*sr)])
plt.savefig('img/i.svg', bbox_inches='tight')
```

```python
# Plot /o/ waveform
plt.figure(figsize=(6, 4), dpi=300)
plt.plot(y[int(1.75*sr):int(1.85*sr)])
plt.savefig('img/o.svg', bbox_inches='tight')
```

```python
# Plot /u/ waveform
plt.figure(figsize=(6, 4), dpi=300)
plt.plot(y[int(3.3*sr):int(3.4*sr)])
plt.savefig('img/u.svg', bbox_inches='tight')
```

```python
# Plot and save spectrogram
y = librosa.effects.preemphasis(y, coef=0.98)
D = librosa.stft(y, n_fft=512, hop_length=16)  # STFT of y
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
min_db = -50
S_db[S_db < min_db] = min_db

plt.figure(figsize=(6, 4), dpi=300)
librosa.display.specshow(S_db, x_axis='time', y_axis='linear', cmap=cm.Greens_r)
plt.ylim([0, 8000])

plt.savefig('img/spectrogram.svg', bbox_inches='tight')
```
